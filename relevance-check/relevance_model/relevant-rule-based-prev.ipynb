```python
nlp = en_core_web_sm.load()
nlp.add_pipe(merge_entities)
```

Statement is relevant is the subject (nsubj or nsubjpass) has policy-related keyword

```python
def is_relevant(sent, keywords):
    doc = nlp(sent)
    for chunk in (doc.noun_chunks):
#         print(chunk, chunk.root.dep_)
#         print(chunk.root.dep_=='nsubj')
        if chunk.root.dep_!='nsubj' and chunk.root.dep_!='nsubjpass':
            continue
        for word in keywords:
            if word in chunk.text:
                return True
    return False 
```

Identify the specific policy

```python
def check_policy(sent, policies):
    p = []
    for policy, keywords in policies.items():
        for word in keywords:
            if word.lower() in sent.lower():
                p.append(policy)
                break
#     if len(p)==0:
#         p.append('tech')
    if len(p)>1:
        print(p)
    return p
```

```python
output = open('out_all.txt','w')
with open('all.txt','r') as file:
    for line in file:
        p = check_policy(line, policies)
        p = " ".join(p)
        output.write(p+"\n")
output.close()
```
