#### Given an economic policy by-statement, labels it as relevant or non-relevant, depending on whether policy-related key-words appear in the subject or object phrase of the sentence

```python
import pandas as pd
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm
from spacy.pipeline import merge_entities
nlp = en_core_web_sm.load()
nlp.add_pipe(merge_entities)
# merge_nps = nlp.create_pipe("merge_noun_chunks")
# nlp.add_pipe(merge_nps)
```

```python
with open("/Users/navreetkaur/MTP/data/relevance_model/manually-labelled-data/relevant_gst.txt",'r') as f:
    sents = f.readlines()
```

```python
for i,sent in enumerate(sents):
    doc = nlp(sent)
    texts = []
    roots = []
    deps = []
    heads = []
    for chunk in doc.noun_chunks:
        texts.append(chunk.text)
        roots.append(chunk.root.text)
        deps.append(chunk.root.dep_)
        heads.append(chunk.root.head.text)
    df = pd.DataFrame({"TEXT":texts, "ROOT.TEXT": roots, "ROOT.DEP_":deps, "ROOOT.HEAD.TEXT":heads})
#     if i==21:
    print(sent)
    print(df)
```

Statement is relevant is the subject (nsubj or nsubjpass) has policy-related keyword

```python
def is_relevant(sent, keywords):
    doc = nlp(sent)
    for chunk in (doc.noun_chunks):
#         print(chunk, chunk.root.dep_)
#         print(chunk.root.dep_=='nsubj')
        if chunk.root.dep_!='nsubj' and chunk.root.dep_!='nsubjpass':
            continue
        for word in keywords:
            if word in chunk.text:
                return True
    return False 
```
